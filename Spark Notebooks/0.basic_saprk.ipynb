```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col


spark = SparkSession.builder \
    .appName("LocalSparkTest") \
    .master("local[*]") \
    .getOrCreate()
```

```python
data = [
    (1,'prashob',40),
    (2,'minnu',35),
    (3,'ram',41),
    (4,'kanchi',35)]
    
```

```python
schema = ('id','name','age')
```

```python
df = spark.createDataFrame(data,schema)
```

```python
df.show()
```

```python
df = df.select('id','name')
```

```python
df.show()
```

```python
df.filter(col('id')>3).show()
```

```python
#wide transformation
# 
df = df.repartition(2)

#narrow

df = df.select('id','name')

df = df.filter(col('id')==1)

df = df.groupBy('id').count()

```
