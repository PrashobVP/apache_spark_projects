```python
from pyspark.sql import SparkSession
from pyspark import pipelines as dp

spark = SparkSession.builder \
    .appName("LocalSparkTest") \
    .master("local[*]") \
    .getOrCreate()

```

```python
path = "../Data/orders.csv"



```

```python

def load_csv():
    df = spark.read.format('csv')\
                .option('inferSchema', True)\
                .option('header',True)\
                .load(path)
    return df
```

```python
df = load_csv()

df.show()
df.count()
df.columns
```
